```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
np.random.seed(1)
data = {
    'x_pos': np.random.randint(0, 20, 100),
    'y_pos': np.random.randint(0, 15, 100),
}

df = pd.DataFrame(data)

heatmap_data = df.groupby(['x_pos', 'y_pos']).size().reset_index(name='visits')

heatmap_pivot = heatmap_data.pivot(index='y_pos', columns='x_pos', values='visits')
heatmap_pivot = heatmap_pivot.fillna(0)

plt.figure(figsize=(10, 5))
sns.heatmap(heatmap_pivot, cmap='YlOrRd')
plt.title('Customer Movement Heatmap (Red = More Visits)')
plt.xlabel('Store Width (X-axis)')
plt.ylabel('Store Depth (Y-axis)')
plt.gca().invert_yaxis()
plt.show()

top_spots = heatmap_data.sort_values(by='visits', ascending=False).head(5)
print("Suggested Product Placement Spots for x & y:")
print(top_spots[['x_pos', 'y_pos', 'visits']])
```
